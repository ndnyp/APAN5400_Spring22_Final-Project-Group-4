### Show image retrieve from Open library api

##### sample api format: 
##### https://covers.openlibrary.org/b/$key/$value-$size.jpg
##### key is the identifier, we use isbn 
##### value is the identifier value
##### size is '-S' (small)

```python
# import necessary library 
from PIL import Image
import requests
from io import BytesIO

# prepare the isbn value and confirm size
isbn = '9781599287829'
file_type = isbn + '-L.jpg'

# putting isbn and file_tye together to form our url
url = 'https://covers.openlibrary.org/b/isbn/' + file_type

# retrieve
response = requests.get(url)
img = Image.open(BytesIO(response.content))

# show 
img

```

```python
import shutil
import requests

# prepare the isbn value and confirm size
isbn = '9781524763138'
file_type = isbn + '-S.jpg'

# putting isbn and file_tye together to form our url
url = 'https://covers.openlibrary.org/b/isbn/' + file_type


response = requests.get(url, stream=True)

# save the image file using isbn as name
save_name = isbn + '.jpg'
with open(save_name, 'wb') as out_file:
    shutil.copyfileobj(response.raw, out_file)
del response
```

```python
from pymongo import MongoClient
client = MongoClient() ## or MongoClient("localhost:27")
db = client.apan5400project
```

```python
# the isbn we are looking for is in the edition collection
edition = db.edition
```

```python
# object = edition.find_one()
result = result = edition.find ({ 'isbn_13' : { '$exists': True } })

# edition.find({ 'isbn_13' : { '$exists': True, '$ne': NULL } });

isbn_list = []

# preparing 1000 isbn for image retrieval
for i in range(1000): 
    isbn_list.append(result[i]['isbn_13'])

```

```python
isbn_list
```

```python
# function to retrieve and download the image locally
def download_image():
    for i in isbn_list:
        # prepare the isbn value and confirm size
        isbn = i[0]
        file_type = isbn + '-L.jpg'

        # putting isbn and file_tye together to form our url
        url = 'https://covers.openlibrary.org/b/isbn/' + file_type


        response = requests.get(url, stream=True)

        # save the image file using isbn as name
        save_name = isbn + '.jpg'
        with open(save_name, 'wb') as out_file:
            shutil.copyfileobj(response.raw, out_file)
        del response
    
```

```python
# run the function, download those pictures locally
download_image()
```

```python
# upload the image to mongodb
import gridfs

#Create a object of GridFs for the above database.
image = gridfs.GridFS(db)

#define an image object with the location.
for i in isbn_list: 
    isbn = i[0]
    file = isbn + '.jpg'

    #Open the image in read-only format.
    with open(file, 'rb') as f:
        contents = f.read()

    #Now store/put the image via GridFs object.
    image.put(contents, filename=file)
```
